```python
%matplotlib inline
import scipy.linalg as splin
import numpy as np
import matplotlib.pyplot as plt
import math
import bisect
import random
import pickle
from mpl_toolkits.mplot3d import proj3d
```

```python
def plot3D(px,py,pz,clist,emphPoints = [],cmap='plasma',shadow=False,useCMAP = True,xLabel='',yLabel='',zLabel='',savefig='',saveN = 1):
    
    elev = 30
    azim = 60

    viewVect = [np.cos(np.pi*180/elev)*np.sin(np.pi*180/azim),np.cos(np.pi*180/elev)*np.cos(np.pi*180/azim),np.cos(np.pi*180/elev)]
    
    thetalist = np.linspace(0,2*np.pi,saveN + 1)
    thetalist = thetalist[:-1]
    
    projMat = np.array([px,py])
    
    for idx in range(saveN):
        theta = thetalist[idx]
        rotMat = np.array([[np.cos(theta),-np.sin(theta)],[np.sin(theta),np.cos(theta)]])
        
        rotx,roty = np.dot(rotMat,projMat)
        rotz = pz
        clistRot = clist
        
        p = np.array([rotx,roty,rotz])
        dots = np.array(np.dot(viewVect,p))
        sort = dots.argsort()

        fig = plt.figure(figsize = (10,10))
        ax = fig.gca(projection='3d')

        Xdist = px.max() - px.min()
        Xcntr = (px.max() + px.min())/2.0
        Ydist = py.max() - py.min()
        Ycntr = (py.max() + py.min())/2.0
        Zdist = pz.max() - pz.min()
        Zcntr = (pz.max() + pz.min())/2.0

        dist = 0.55*max(Xdist,Ydist,Zdist)

        for p in emphPoints:
            ax.scatter([rotx[p]],[roty[p]],[rotz[p]],c='C1',s=200)
            if shadow:
                ax.scatter([rotx[p]],[roty[p]],[Zcntr - dist],c='black',alpha=0.5)
                ax.scatter([rotx[p]],[Ycntr + dist],[rotz[p]],c='black',alpha=0.5)
                ax.scatter([Xcntr - dist],[roty[p]],[rotz[p]],c='black',alpha=0.5)
        
        rotx = rotx[sort]
        roty = roty[sort]
        rotz = rotz[sort]
        clistRot = np.array(clistRot)[sort]
        
        if useCMAP:
            ax.scatter(rotx,roty,rotz,c=clistRot,s=100,edgecolor='black',linewidth=0.0,cmap = cmap)
        else:
            ax.scatter(rotx,roty,rotz,c=clistRot,s=100,edgecolor='black',linewidth=0.0)


        ax.set_xlim([Xcntr - dist, Xcntr + dist])
        ax.set_ylim([Ycntr - dist, Ycntr + dist])
        ax.set_zlim([Zcntr - dist, Zcntr + dist])

        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_zticks([])

        #ax.set_axis_off()
        #plt.show()

        if shadow:
            #fig = plt.figure(figsize = (10,10))
            #ax = fig.gca(projection='3d')
            #ax.view_init(30,-60)

            ax.scatter(rotx,roty,[Zcntr - dist for p in rotz],c='black',alpha=0.01)
            ax.scatter(rotx,[Ycntr + dist for p in roty],rotz,c='black',alpha=0.01)
            ax.scatter([Xcntr - dist for p in rotx],roty,rotz,c='black',alpha=0.01)

            ax.set_xlim([Xcntr - dist, Xcntr + dist])
            ax.set_ylim([Ycntr - dist, Ycntr + dist])
            ax.set_zlim([Zcntr - dist, Zcntr + dist])

            ax.set_xticks([])
            ax.set_yticks([])
            ax.set_zticks([])

            #ax.set_axis_off()
        ax.set_xlabel(xLabel,fontsize=16)
        ax.set_ylabel(yLabel,fontsize=16)
        ax.set_zlabel(zLabel,fontsize=16)

        if savefig == '':
            plt.show()
        else:
            plt.savefig(savefig+str(idx)+'.png')
            plt.close()
```

```python
def saveThing(thing,file):
    with open(file,'wb') as fil:
        pick = pickle.Pickler(fil,protocol=4)
        pick.dump(thing)
        pick.clear_memo()
```

```python
def FIM(x):
    return 16/np.pi * np.cos(x)**2 * np.sin(x)**2

def cdfFIM(x):
    return (4*x - np.sin(4*x))/(2*np.pi)
```

```python
def JPrior(x):
    return 2 * np.cos(x) * np.sin(x)

def cdfJPrior(x):
    return np.sin(x)**2
```

```python
thetas = np.linspace(0.0,np.pi/2.0,2001)
weights = cdfJPrior(thetas)
```

```python
points = [np.pi/4,0]
for i in range(4000):
    x = random.random()
    idx = bisect.bisect(weights,x)
    points.append(thetas[idx])
```

```python
n,bins,patches = plt.hist(points,normed=1,bins=50)
plt.plot(thetas,JPrior(thetas))
plt.show()
```

```python
plt.figure(figsize=(4,4))

plt.scatter(np.cos(points)**2,np.sin(points)**2,c=points)

plt.scatter([np.cos(points[0])**2],[np.sin(points[0])**2],c='C1')

plt.xlabel('Probability of Heads',fontsize=14)
plt.ylabel('Probability of Tails',fontsize=14)

#plt.title('Rigged Coin Toss',fontsize=16)

plt.show()
```

```python
DDT = [[-4*np.log(np.cos(t1-t2)) for t1 in points] for t2 in points]
sums = [sum(row)/len(points) for row in DDT]
sum_sums = sum(sums)/len(points)
CC = [[DDT[i][j] - sums[i] - sums[j] + sum_sums for i in range(len(DDT))] for j in range(len(DDT))]
```

```python
w,v = np.linalg.eig(CC)
```

```python
proj = np.dot(v,np.diag(np.sqrt(abs(w)))).real
```

```python
plt.figure(figsize=(8,8))
distances = np.array([max(proj[:,i])-min(proj[:,i]) for i in range(len(w))])
selection = [i for i in range(0,100)]
y = distances[selection]
x = abs(w).real[selection]
plt.plot(x,y,marker='o')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
plt.figure(figsize=(4,4))
x = proj[:,0]
y = proj[:,1]
plt.scatter(x,y)
fairIdx = 0
plt.scatter([x[fairIdx]],[y[fairIdx]],c='C1')
plt.axes().set_aspect('equal', 'datalim')

plt.xlabel('First Component\n(Positive Squared Distance)')
plt.ylabel('Second Component\n(Negative Squared Distance)')

plt.title('IPCA of Coin Toss')

plt.show()
```

```python
px = proj[:,0]
py = proj[:,1]
pz = proj[:,2]

clist = points

plot3D(px,py,pz,clist,emphPoints = [0],shadow=True,xLabel='Positive Squared Distance',yLabel='Negative Squared Distance',zLabel='Positive Squared Distance')
```

```python
imCols = []
reCols = []
for i in range(len(w)):
    if w[i].real < 0:
        imCols.append(i)
    else:
        reCols.append(i)
```

```python
fairProj = proj[0]
```

```python
dr2 = []
di2 = []
for p in proj:
    ddif2 = (fairProj - p)**2
    dr2_temp = sum([ddif2[x] for x in imCols])
    di2_temp = sum([ddif2[x] for x in reCols])
    dr2.append(dr2_temp)
    di2.append(di2_temp)
dr2 = np.array(dr2)
di2 = np.array(di2)
```

```python
plt.figure(figsize=(10,8))
plt.plot(thetas,[-8*np.log(np.cos(points[0] - t)) for t in thetas],c='black',label = 'Intensive Distance')
plt.scatter(points,dr2-di2, label = 'Total Distance from IPCA')
plt.scatter(points,dr2,label='Real Distances from IPCA')
plt.scatter(points,-di2,label = 'Imaginary Distances from IPCA')
#plt.xticks([i*np.pi/8 for i in [0,1,2,3,4]])
#plt.yticks([-10,0,10])
#plt.xlim([min(points),max(points)])
#plt.ylim([-10,10])
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.plot(np.cos(thetas)**2,[-8*np.log(np.cos(points[0] - t)) for t in thetas],c='black',label = 'Intensive Distance')
plt.scatter(np.cos(points)**2,dr2-di2, label = 'Total Distance from IPCA',s=10)
plt.scatter(np.cos(points)**2,dr2,label='Positive Squared Distances',s=5)
plt.scatter(np.cos(points)**2,-di2,label = 'Negative Squared Distances',s=5)

plt.xlabel('Probability of Heads',fontsize=14)
plt.ylabel('Distance from Fair Coin',fontsize=14)

plt.ylim([-5,5])

plt.xticks([0,0.25,0.5,0.75,1],['0','1/4','Fair Coin','3/4','1'])

plt.grid()
plt.legend()
plt.show()
```
